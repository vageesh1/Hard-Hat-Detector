Function for getting the dominant color

```python
def unique_count_app(a):
    '''
    a: NumPy array i.e. the ROI from your image
    Returns: tuple (r, g, b)
    '''
    colors, count = np.unique(
        a.reshape(-1, a.shape[-1]), axis=0, return_counts=True)
    return colors[count.argmax()]
```

```python
def run_on_single_image(img):
  img_copy=img.copy()
  img_copy = cv2.cvtColor(img_copy, cv2.COLOR_BGR2RGB)
  output = model(img_copy)
  preds=output.pandas().xyxy[0] 
  if preds is not None:
    for i in range(0,len(preds)):
      if(preds.iloc[i]['name']=='helmet'):
        xmin=int(preds.iloc[i]['xmin'])
        ymin=int(preds.iloc[i]['ymin'])
        xmax=int(preds.iloc[i]['xmax'])
        ymax=int(preds.iloc[i]['ymax'])
        #adjusting ymin and ymax
        crop_img=img[ymin+5:ymax-15,xmin:xmax]
        #reshaping it 
        data = np.reshape(crop_img, (-1,3))
        if data is not None:
          dom_color=unique_count_app(crop_img).tolist
        #creating the new image 
          c1=dom_color()[0]
          c2=dom_color()[1]
          c3=dom_color()[2]
          img=cv2.rectangle(img,(xmin,ymin),(xmax,ymax),(c1,c2,c3),2)
          cv2.putText(img, 'Helmet', (xmin, ymin-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (c1,c2,c3), 2)
          cv2.putText(img, str(int(preds.iloc[i]['confidence']*100)), (xmin+100, ymin-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (c1,c2,c3), 2)
        else:
          continue

      #for head I am taking green as the color
      elif(preds.iloc[i]['name']=='head'):
        xmin=int(preds.iloc[i]['xmin'])
        ymin=int(preds.iloc[i]['ymin'])
        xmax=int(preds.iloc[i]['xmax'])
        ymax=int(preds.iloc[i]['ymax'])
        img=cv2.rectangle(img,(xmin,ymin),(xmax,ymax),(0,255,0),2)
        cv2.putText(img, 'Head', (xmin, ymin-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)
        cv2.putText(img, str(int(preds.iloc[i]['confidence']*100)), (xmin+100, ymin-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (0,255,0), 2)

      #for person i am taking yellow as the color

      elif(preds.iloc[i]['name']=='person'):
        xmin=int(preds.iloc[i]['xmin'])
        ymin=int(preds.iloc[i]['ymin'])
        xmax=int(preds.iloc[i]['xmax'])
        ymax=int(preds.iloc[i]['ymax'])
        img=cv2.rectangle(img,(xmin,ymin),(xmax,ymax),(0,255,0),2)
        cv2.putText(img, 'Person', (xmin, ymin-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255,255,0), 2)
        cv2.putText(img, str(int(preds.iloc[i]['confidence'*100])), (xmin+100, ymin-10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255,255,0), 2)
      #if none are there then continue 
      else:
        continue
      
  return img
```

Now after getting the results in JSOnN we can convert them to teh XML files using library CSV Pascal to VOC

```python
# Convert CSV file to XML string
import csv
filename = '/content/drive/MyDrive/datasets/hard_hat(yolov5)/test data/preds/hard_hat_workers1068.csv'
def convert_row(headers, row):
    s = f'<row id="{row[0]}">\n'
    for header, item in zip(headers, row):
        s += f'    <{header}>' + f'{item}' + f'</{header}>\n'
    return s + '</row>'
with open(filename, 'r') as f:
    r = csv.reader(f)
    headers = next(r)
    xml = '<data>\n'
    for row in r:
        xml += convert_row(headers, row) + '\n'
    xml += '</data>'
    print(xml)
```

```python
def json2xml(json_obj, line_padding=""):
    result_list = list()

    json_obj_type = type(json_obj)

    if json_obj_type is list:
        for sub_elem in json_obj:
            result_list.append(json2xml(sub_elem, line_padding))

        return "\n".join(result_list)

    if json_obj_type is dict:
        for tag_name in json_obj:
            sub_obj = json_obj[tag_name]
            result_list.append("%s<%s>" % (line_padding, tag_name))
            result_list.append(json2xml(sub_obj, "\t" + line_padding))
            result_list.append("%s</%s>" % (line_padding, tag_name))

        return "\n".join(result_list)

    return "%s%s" % (line_padding, json_obj)
```
